```python
import csv
import numpy as np 
import matplotlib.pyplot as plt 

f= open('age.csv', encoding='cp949')
data = csv.reader(f)

header= next(data)

data= list(data)
min_val=1
result_name=''
result=0

name=input('  :')
for row in data:
    for i in range(102):
        row[i+2] = int(row[i+2].replace(',',""))
    if '' in row[0]:
        home = np.array(row[3:], dtype=int)/row[2]

for row in data:
    if name not in row[0]:
        away =np.array(row[3:], dtype=int)/row[2]
        s= np.sum((home-away)**2)
        if s< min_val:
            nim_val=s
            result_name = row[0]
            result=away

f.close()

plt.rc('font', family='Malgun Gothic')
plt.title(name+ ' :'+ result_name)
plt.grid(True)
plt.xlabel('')
plt.ylabel('')
plt.plot(home, label=name)
plt.plot(result, label=result_name)
plt.legend()
plt.show()
```
