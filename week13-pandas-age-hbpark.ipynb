```python
df = pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84')
df2 = df[0].set_index('국가 (IOC 코드)')

summer=df2.iloc[:, :5]
# iloc는 integer location의 약자, 데이터프레임의 행이나 열의 순서를 나타내는 정수로 특정값 추출
# : -> 슬라이스 이므로 미만
summer
```

```python
df = pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84')
df2 = df[0].set_index('국가 (IOC 코드)')

summer=df2.iloc[:, :5]

summer.columns=['경수기','금','은','동','합계']
#columns에 이름 설정
summer
```

```python
df = pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84')
df2 = df[0].set_index('국가 (IOC 코드)')

summer=df2.iloc[:, :5]

summer.columns=['경수기','금','은','동','합계']
summer = summer.sort_values('금',ascending=False)
# sort_values() 함수로 원하는 열을 기준으로 데이터순서 정렬
# ascending = 오름차순
summer
```

```python
df = pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84')
df2 = df[0].set_index('국가 (IOC 코드)')

summer=df2.iloc[:, :5]

summer.columns=['경수기','금','은','동','합계']
summer = summer.sort_values('금',ascending=False)
summer.to_excel('하계 올림픽메달.xlsx')
```
