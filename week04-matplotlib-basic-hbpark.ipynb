```python
import matplotlib.pyplot as plt


plt.plot([1,2,3,4],[10,20,30,40], label="skyblue", color='skyblue')
plt.plot([1,2,3,4],[40,30,20,10], 'pink',label="pink")

plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.plot([1,2,3,4],[10,20,30,40], color="r", linestyle='--',label='dashed')
plt.plot([1,2,3,4],[5,10,30,20], 'y',ls='-.',label="dash-dot")
plt.plot([1,2,3,4],[40,30,20,10], 'b',ls=':',label="dotted")

plt.title('linestyle')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.plot([1,2,3,4],[10,20,30,40], color="r", linestyle='--', marker='D', label='diamond')
plt.plot([1,2,3,4],[5,10,30,20], 'y', ls='-.',marker='^', label="triangle")
plt.plot([1,2,3,4],[40,30,20,10], 'b', ls=':', marker='*', label="star")

plt.title('marker')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.plot([1,2,3,4],[10,20,30,40], color='skyblue',linewidth=3, linestyle='--', marker='*',label="skyblue")
plt.plot([1,2,3,4],[40,30,20,10], 'pink',linewidth=1, ls=':', marker='D', label="pink")

plt.title('linewidth')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0,10, N, endpoint=True)
x2 = np.linspace(0,10, N, endpoint=False)
plt.plot(x1, y + 0.5,'o', label='endpoint=True')
plt.plot(x2, y, 'o', label='endpoint=False')

plt.legend()
plt.ylim([-0.5,1])
plt.show()
```
